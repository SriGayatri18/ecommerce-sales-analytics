```python
import pandas as pd

url = "https://raw.githubusercontent.com/SriGayatri18/ecommerce-sales-analytics/main/data_raw/Sample%20-%20Superstore.csv"
df = pd.read_csv(url, encoding='latin1')

df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
df.duplicated().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Ship Delay (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100
```

```python
df.to_csv("superstore_sales_cleaned.csv", index=False)
```
